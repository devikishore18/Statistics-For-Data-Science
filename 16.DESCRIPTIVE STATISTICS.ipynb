## What is Descriptive Statistics?

Descriptive statistics help summarize and describe the **main features** of a dataset using:

- **Central Tendency**: Mean, Median, Mode  
- **Spread (Dispersion)**: Standard Deviation, Variance, Range, IQR  
- **Shape**: Skewness, Kurtosis  
- **Distribution overview**: Min, Max, Count, Percentiles

---



## ðŸ“¦ Libraries to Use



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## SAMPLE DATASET

```python
data = {
    'Marks': [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
}
df = pd.DataFrame(data)
```

## 1.CENTRAL TENDENCY
Mean is the average, Median is the middle value, Mode is the most frequent.



```python
print("Mean:", df['Marks'].mean())
print("Median:", df['Marks'].median())
print("Mode:", df['Marks'].mode()[0])
```

## 2. Dispersion (Spread)
Shows how spread out the data is. Standard deviation is in the same unit as data.



```python
print("Standard Deviation:", df['Marks'].std())
print("Variance:", df['Marks'].var())
print("Range:", df['Marks'].max() - df['Marks'].min())
```

## 3. Percentiles and Quartiles
Useful to detect outliers and understand data distribution.




```python
print("25th Percentile (Q1):", df['Marks'].quantile(0.25))
print("50th Percentile (Q2 - Median):", df['Marks'].quantile(0.5))
print("75th Percentile (Q3):", df['Marks'].quantile(0.75))

# Interquartile Range
iqr = df['Marks'].quantile(0.75) - df['Marks'].quantile(0.25)
print("Interquartile Range (IQR):", iqr)
```

## 4. Summary Using .describe()
Gives count, mean, std, min, Q1, Q2 (median), Q3, max.



```python
df['Marks'].describe()
```

## 5. Skewness & Kurtosis
Skewness tells if data is left/right skewed.

Kurtosis tells if data has heavy/light tails.



```python
print("Skewness:", df['Marks'].skew())
print("Kurtosis:", df['Marks'].kurt())
```

## 6. Visualization (Hist, Box, Violin)



```python
# Histogram
sns.histplot(df['Marks'], kde=True)
plt.title("Histogram of Marks")
plt.show()

# Boxplot
sns.boxplot(y=df['Marks'])
plt.title("Boxplot of Marks")
plt.show()

# Violin Plot
sns.violinplot(y=df['Marks'])
plt.title("Violin Plot of Marks")
plt.show()
```
